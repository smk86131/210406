```python
import random
correct_answer = random.randint(1, 20)

while True :
    number = int(input("숫자를 입력하세요: "))

    if correct_answer == number :
        break
    elif correct_answer > number :
        print(number, "보다 큽니다")
    else :
        print(number, "보다 작습니다")

print("정답입니다")
```
